```python
1: Load Cleaned Data
import pandas as pd

df = pd.read_csv("../data/processed/student_performance_cleaned.csv")
df.head()

Observation: The cleaned dataset is loaded successfully for analysis.

2: Target Variable Distribution
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Final_Marks"], bins=10, kde=True)
plt.title("Final Marks Distribution")
plt.show()

Observation: Final marks are approximately normally distributed with few outliers.

3: Attendance Analysis
sns.histplot(df["Attendance"], bins=10, kde=True)
plt.title("Attendance Distribution")
plt.show()

Observation: Most students have attendance above 70%, indicating moderate discipline.

4: Gender vs Performance
sns.boxplot(x="Gender", y="Final_Marks", data=df)
plt.title("Marks by Gender")
plt.show()

Observation: Female students show slightly higher median performance.

5: Attendance vs Marks
sns.scatterplot(x="Attendance", y="Final_Marks", data=df)
plt.title("Attendance vs Final Marks")
plt.show()

Observation: Higher attendance generally leads to better academic performance.

6: Study Hours vs Marks
sns.scatterplot(x="Study_Hours", y="Final_Marks", data=df)
plt.title("Study Hours vs Final Marks")
plt.show()

Observation: Marks increase with study hours up to a certain limit.

7: Correlation Matrix
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

Attendance and study hours show strong positive correlation with final marks.
```
